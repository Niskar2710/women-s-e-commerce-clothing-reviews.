# WOMEN'S E-CLOTHING CUSTOMER REVIEWS:

**Introduction:**
E-commerce has transformed shopping, especially in fashion, with women's clothing being a dominant category. Reliable reviews are crucial for consumers and businesses, as they help shape consumer preferences and purchasing behaviors, ultimately improving offerings and customer satisfaction.


**Project objective:**
This project aims to analyze women's e-commerce clothing reviews, focusing on product quality, fit, pricing, customer service, and satisfaction. It seeks to identify trends and sentiment expressions, providing insights for consumers and businesses. The research contributes to understanding the role of reviews in shaping women's online clothing shopping experiences.
Dataset Selection: For this analysis, we will utilize a dataset containing information about " Women's E-Commerce Clothing Reviews".
1.	Informed Purchasing Decisions: Women's e-commerce clothing reviews serve as a vital source of information for potential buyers, helping them make informed decisions about the quality, sizing, style, and overall satisfaction with a particular clothing item before making a purchase.
2.	Trust and Credibility: Reviews from other women who have purchased and used the clothing items provide a sense of trust and credibility, especially in an online shopping environment where physical inspection of the product is not possible. Positive reviews can instill confidence in potential buyers, while negative reviews can alert them to potential issues or drawbacks.
3.	Feedback for Improvement: For e-commerce businesses, women's clothing reviews offer valuable feedback on product design, quality, sizing accuracy, shipping experience, and customer service. Analyzing these reviews can help businesses identify areas for improvement and refine their offerings to better meet customer needs and preferences.
4.	Community Engagement: Reviews create a sense of community among shoppers, allowing them to share their experiences, recommendations, and tips with others. This engagement fosters a loyal customer base and encourages repeat purchases.
5.	SEO and Visibility: Positive reviews can improve a product's visibility and search engine ranking on e-commerce platforms, as algorithms often prioritize items with higher ratings and more reviews. This increased visibility can lead to more traffic and sales for businesses.
6.	Market Research: Analyzing women's e-commerce clothing reviews provides valuable insights into consumer preferences, trends, and demands within the fashion industry. This data can inform product development, marketing strategies, and inventory management decisions.
7.	Competitive Analysis: Studying reviews of competitors' products allows businesses to benchmark themselves against industry standards, identify competitive advantages, and differentiate their offerings to attract more customers.
8.	Brand Reputation Management: Women's e-commerce clothing reviews play a significant role in shaping brand reputation and perception. Monitoring and responding to reviews, both positive and negative, demonstrate a commitment to customer satisfaction and can help mitigate potential damage to a brand's reputation.
By understanding these purposes and conducting thorough analysis of women's e-commerce clothing reviews, the project aims to uncover actionable insights that benefit both consumers and businesses in the online fashion marketplace.

**DATA VISUALIZATION**

**Rating on basis of Clothing ID:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_e-commerce.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(4)
customerDf['Clothing ID'].groupby(customerDf['Rating']).sum().plot(kind='bar')

plt.ylabel('Clothing ID')
plt.xlabel('Rating')
plt.title('Rating on basis of Clothing ID')

plt.show()
```

**Recommended IND:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_e-commerce.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure(figsize=(9,5))
mycolor=np.array(['green','pink'])
customerDf['Recommended IND'].value_counts().plot(kind='pie',autopct='%1.0f%%',colors=mycolor)
plt.title('Recommended IND')
plt.legend(title='Layouts')
plt.show()
```

**Rating according to division name:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_e-commerce.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(5)
mycolor=np.array(['pink','lightgreen','skyblue'])
customerDf['Rating'].groupby(customerDf['Division Name']).sum().plot(kind='bar', color=mycolor)
plt.xlabel('Division Name')
plt.ylabel('Rating')
plt.title('Rating according to division name')
plt.show()
```

**Departmentwise Recommended IND:**

```python
dataset=pd.read_csv('after_drop_e-commerce.csv')
colors = ['blue', 'yellow','red']
dataset.groupby('Department Name')['Recommended IND'].sum().sort_values(ascending=False).head(10).plot(kind='barh',xlabel='Recommended IND',ylabel='Department Name',title='Departmentwise Recommended IND',color=colors)
```

**Departmentwise Recommended IND:**

```python
dataset=pd.read_csv('after_drop_e-commerce.csv')
colors = ['cyan', 'gold','purple','orange','green','maroon','pink']
dataset.groupby('Class Name')['Positive Feedback Count'].sum().sort_values(ascending=False).head(10).plot(kind='barh',xlabel='Positive Feedback Count',ylabel='class Name',title='Departmentwise Recommended IND',color=colors)
```

**Reviews based on rating:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_e-commerce.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure(figsize=(9,5))
mycolor=np.array(['orange','yellow','grey','pink','royalblue'])
customerDf['Rating'].value_counts().plot(kind='pie',autopct='%1.0f%%',colors=mycolor)
plt.title('Reviews based on rating')
plt.legend(title='Rating layouts')
plt.show()
```

**Reviews based on rating:**

```python
dataset=pd.read_csv('after_drop_e-commerce.csv')
colors = ['magenta', 'red','lightgreen','blue','skyblue','orange','brown']
dataset.groupby('Title')['Rating'].sum().sort_values(ascending=False).head(10).plot(kind='barh',xlabel='Rating',ylabel='Title',title='Customer reviews correspond to rating',color=colors)
```

**Customer reviews based on age:**

```python
dataset=pd.read_csv('after_drop_e-commerce.csv')
colors = ['orange', 'cyan','green','pink','grey','yellow','magenta']
dataset.groupby('Review Text')['Age'].sum().sort_values(ascending=False).head(10).plot(kind='barh',xlabel='Age',ylabel='Review Text',title='Customer reviews based on age',color=colors)
```
